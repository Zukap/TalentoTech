```python
import keras
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# Generar un conjunto de datos de clasificacion sintetico
X, y = make_classification (n_samples=100, n_features=5, n_classes=2)
# Dividir los estados en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Convertir las etiquetas en conjuntos de entrenamienot y prueba
y_train_categorical =keras.utils.to_categorical(y_train)
y_test_categorical = keras.utils.to_categorical(y_test)
```

```python
model = keras.models.Sequential()
#Anadir una capa de salida con 2 neuronas (una para cada clase) y activacion softmax
model.add(keras.layers.Dense(10, input_shape=(X_train.shape[1],), activation='relu'))

model.add(keras.layers.Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train_categorical, epochs=10, batch_size=32, validation_data= (X_test, y_test_categorical))
```

```python
loss, accuracy = model.evaluate(X_test, y_test_categorical)
print("Perdida en el conjunto de prueba",loss)
print("Perdida en el conjunto de prueba",accuracy)
```
